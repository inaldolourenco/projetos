```python
# baixa/grava as fotos dos deputados no pc C:\Users\Usuario\Untitled Folder
import requests
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados?siglaUf=&siglaUf=&ordem=ASC&ordenarPor=nome'
resp = requests.get(url).json()
for d in resp['dados']:
    if d['siglaPartido'] == 'PSL' and d['siglaUf'] == 'SP': #sigla do partido e sigla do Estado
    nome = d['nome'].lower()   
    print (f'Gravando: {nome}')
    f = open(nome + '.jpg', 'wb')
    foto = requests.get(d['urlFoto']).content # pega o conteudo
    f.write(foto) # grava
    f.close() #fecha conexa
```

```python
# Despesa por deputado
import requests
total = 0
for pag in range (1,100): # pega a pagina {pag} no intervalo de 1 a 100
    u = f'https://dadosabertos.camara.leg.br/api/v2/deputados/107283/despesas?ano=2019&ordem=ASC&ordenarPor=ano&pagina={pag}&itens=100'
    r = requests.get(u).json()
    for gasto in r['dados']:
        total = total + float(gasto ['valorLiquido'])  # converte o valor texto em float
        print(f'Gasto em 2019: R$ {total:.2f}')
```
